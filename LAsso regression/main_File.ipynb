# Lasso Regression:
    Y = wX + b
    Y --> Dependent Variable
    X --> Independent Variable
    w --> weight
    b --> bias


**Gradient Descent:**


Gradient Descent is an optimization algorithm used for minimizing the loss function in various machine learning algorithms. It is used for updating the parameters of the learning model.


w  =  w - α*dw


b  =  b - α*db



**Learning Rate:**


Learning rate is a tuning parameter in an optimization algorithm that determines the step size at each iteration while moving toward a minimum of a loss function.


```python
import pandas as pd
```

```python
df = pd.read_csv("salary.csv")
df.head()
```

```python
from app2 import LassoRegression
from sklearn.linear_model import Lasso
from sklearn.metrics import accuracy_score
```

```python
model1 = Lasso()
model2 = LassoRegression(num_iterations=1000,learning_rate=0.01,lambda_param=1)
```

```python
from sklearn.model_selection import train_test_split

X  = df.drop("Salary",axis=1)
Y  = df[["Salary"]]

X_train , X_test ,Y_train , Y_test = train_test_split(X,Y,test_size=0.25)
```

```python
print(X_train.shape)
print(Y_train.shape)
```

```python
Y_train = Y_train.values.flatten()
Y_test = Y_test.values.flatten()
```

```python
model2.fit(X_train,Y_train)
```

```python
Y_pred_model2 = model2.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error ,r2_score


# Calculate evaluation metrics
mse = mean_squared_error(Y_test, Y_pred_model2)
mae = mean_absolute_error(Y_test, Y_pred_model2)
r_score = r2_score(Y_test,Y_pred_model2)

# Print the evaluation metrics
print("Mean Squared Error (MSE):", mse)
print("Mean Absolute Error (MAE):", mae)
print("R-squared:", r_score)
```
